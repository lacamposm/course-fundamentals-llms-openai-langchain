## ***La ingeniería de prompts.***

La ingeniería de prompts se refiere al proceso de diseñar y formular preguntas o instrucciones específicas para obtener respuestas deseadas de modelos de lenguaje como los LLMs. En este contexto, se busca manipular la salida del modelo mediante la entrada dada, aprovechando su capacidad de generar texto coherente.

### ***Componentes clave de la ingeniería de prompts:***

1. **Formulación precisa:**

   - Seleccionar palabras y estructuras que guíen al modelo hacia la respuesta deseada.
   - Considerar la sensibilidad del modelo a ciertos términos y expresiones.

2. **Ajuste de contexto:**

   - Introducir información contextual para orientar al modelo.
   - Experimentar con la longitud y complejidad del contexto para influir en la respuesta.

3. **Iteración y experimentación:**

   - Probar diferentes enfoques y ajustes en los prompts.
   - Analizar la salida del modelo y ajustar según sea necesario.

4. **Control de sesgo:**

   - Identificar y gestionar sesgos inherentes en los modelos.
   - Modificar los prompts para minimizar resultados no deseados.

5. **Adaptación al dominio:**

   - Personalizar los prompts para adecuarse al dominio específico de la tarea.
   - Incorporar términos técnicos o jerga relevante.
